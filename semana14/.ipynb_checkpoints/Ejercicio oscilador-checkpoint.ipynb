```python
#opcion 1 para definir funciones:
fx=lambda x: x**2*np.cos(x)-3*x**3
```

```python
#opcion 2 para definir funciones
def fx2(x,y): 
    z=x*y*np.cos(x/y)-3*x**3
    return z
```

```python
fx2
```

```python
fx2(3,4)
```

```python
plt.plot(x,fx2(x,y))
```

```python
plt.plot(x,fx2(y,x))
```

```python
fx2(2,4)
```

```python
x3=np.random.rand(10)
print(fx2(x3,2))
plt.plot(fx2(x3,2))
```

```python
def recta(a,m=1,b=0,label='x'):
    return m*a+b
```

```python
x=np.linspace(-5,5,20)
plt.plot(x,recta(x,3))
plt.plot(x,recta(x,3,-2))
plt.plot(x,recta(x,-3,0))
plt.plot(x,recta(x,-2,3))
plt.grid()
```
